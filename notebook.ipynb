## Creating Embeddings

```python
## Creating embeddings from huggingface inference api
import requests

API_URL = "https://api-inference.huggingface.co/models/maidalun1020/bce-embedding-base_v1"
headers = {f"Authorization": f"Bearer {HUGGINGFACE_TOKEN}"}


def query(payload):
	response = requests.post(API_URL, headers=headers, json=payload)
	return response.json()
	
output = query({
	"inputs": "Today is a sunny day and I will get some ice cream.",
})

print(output)
```

```python
def calculate_total_cost(cost_per_million, total_tokens):
    cost_per_token = cost_per_million / 1000000
    total_cost = cost_per_token * total_tokens
    return total_cost
```
